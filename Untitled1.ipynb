```python
from datetime import datetime, timedelta

# Define a function to calculate the time difference between two timestamps
def calculate_time_difference(start_time, end_time):
    fmt = '%m/%d/%Y %I:%M %p'
    start = datetime.strptime(start_time, fmt)
    end = datetime.strptime(end_time, fmt)
    return (end - start).total_seconds() / 3600  # Convert seconds to hours

# Read data from the file
with open('/content/Assignment_Timecard.xlsx - Sheet1.csv', 'r') as file:
    lines = file.readlines()

# Initialize variables to keep track of employees' data
employee_data = {}
current_employee = None
for line in lines[1:]:  # Skip the header
    columns = line.split('\t')
    if len(columns) < 8:
        continue  # Skip lines with insufficient columns

    position_id = columns[0].strip()
    status = columns[1].strip()
    start_time = columns[2].strip()
    end_time = columns[3].strip()
    employee_name = columns[7].strip()

    if status == 'Active':
        if current_employee is None:
            current_employee = {
                'name': employee_name,
                'start_time': start_time,
                'end_time': end_time,
                'days_worked': {start_time.split()[0]},
                'hours_worked': calculate_time_difference(start_time, end_time),
            }
        else:
            if employee_name == current_employee['name']:
                current_employee['end_time'] = end_time
                current_employee['days_worked'].add(start_time.split()[0])
                current_employee['hours_worked'] += calculate_time_difference(
                    start_time, end_time
                )
            else:
                employee_data[current_employee['name']] = current_employee
                current_employee = {
                    'name': employee_name,
                    'start_time': start_time,
                    'end_time': end_time,
                    'days_worked': {start_time.split()[0]},
                    'hours_worked': calculate_time_difference(start_time, end_time),
                }

# Check the conditions for each employee
for name, data in employee_data.items():
    consecutive_days_worked = len(data['days_worked']) >= 7
    less_than_10_hours_between_shifts = (
        data['hours_worked'] >= 1 and data['hours_worked'] < 10
    )
    more_than_14_hours_in_single_shift = data['hours_worked'] > 14
    if consecutive_days_worked or less_than_10_hours_between_shifts or more_than_14_hours_in_single_shift:
        print(f"Employee Name: {name}")
        print(f"Position ID: {position_id}")
        print("Conditions met:")
        if consecutive_days_worked:
            print("  - Worked for 7 consecutive days")
        if less_than_10_hours_between_shifts:
            print("  - Less than 10 hours between shifts")
        if more_than_14_hours_in_single_shift:
            print("  - Worked for more than 14 hours in a single shift")
        print("=" * 40)
```
